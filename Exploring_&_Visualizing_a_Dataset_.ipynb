```python
import seaborn as sns   # Import Seaborn for advanced visualizations
import pandas as pd

df = sns.load_dataset("iris")
df.to_csv("iris.csv", index=False)
```

```python
#Display the first 5 rows of the DataFrame to get a quick look at the data.
df.head()
```

```python
# Show concise summary of DataFrame: column types and non-null counts
df.info()
```

```python
 # Generate descriptive statistics for numeric columns
df.describe()
```

```python
  # Import Matplotlib for creating plots
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(           # Scatter plot of Sepal vs Petal Length colored by species
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Scatterplot of Sepal Length vs petal length")     # Set title of the plot
plt.show()   # Display the scatter plot
```

```python
sns.scatterplot(data=df, x="sepal_width", y="petal_width", hue="species")  # Scatter plot of Sepal Width vs Petal Width colored by species
plt.title("Scatterplot of Sepal width vs petal Width")   # Set title of the plot
plt.show()   # Display the scatter plot
```

```python
plt.hist(df["petal_length"], bins=20)   # Histogram of petal length with 20 bins
plt.title("Distribution of Petal Length")   # Set title
plt.xlabel("Petal Length")   # X-axis label
plt.ylabel("Frequency")   # Y-axis label
plt.show()   # Display the histogram
```

```python
plt.hist(df["sepal_length"], bins=20)   # Histogram of sepal length with 20 bins
plt.title("Distribution of Sepal Length")   # Set title
plt.xlabel("Sepal Length")   # X-axis label
plt.ylabel("Frequency")   # Y-axis label
plt.show()   # Display the histogram
```

```python
plt.hist(df["petal_width"], bins=20)   # Histogram of petal width with 20 bins
plt.title("Distribution of Petal Width")   # Set title
plt.xlabel("Petal Width")   # X-axis label
plt.ylabel("Frequency")   # Y-axis label
plt.show()   # Display the histogram
```

```python
sns.boxplot(x="species", y="sepal_length", data=df)   # Box plot of sepal length grouped by species
plt.title("Box Plot of Sepal Length by Species")   # Set title
plt.show()   # Display the box plot
```

```python
sns.boxplot(x="species", y="petal_length", data=df)   # Box plot of petal length grouped by species
plt.title("Box Plot of Petal Length by Species")   # Set title
plt.show()   # Display the box plot
```

```python
sns.boxplot(
    x="species",
    y="petal_length",
    data=df
)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```
