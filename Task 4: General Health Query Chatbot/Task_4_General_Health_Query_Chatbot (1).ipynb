```python
!pip install requests -q

import requests

API_TOKEN = "hagging_face_token"

MODEL_URL = "https://api-inference.huggingface.co/models/google/flan-t5-base"

headers = {
    "Authorization": f"Bearer {API_TOKEN}"
}

def safety_filter(user_input):
    emergency_keywords = [
        "suicide",
        "kill myself",
        "overdose",
        "chest pain",
        "heart attack",
        "stroke",
        "can't breathe",
        "severe bleeding"
    ]

    for word in emergency_keywords:
        if word in user_input.lower():
            return True
    return False


def create_prompt(user_question):
    system_prompt = """
You are a helpful and friendly medical assistant chatbot.
Provide general health information only.
Do NOT give diagnosis or prescribe medication.
Always suggest consulting a qualified doctor.
Keep answers simple and safe.
"""
    return system_prompt + "\nQuestion: " + user_question


def generate_response(prompt):

    payload = {
        "inputs": prompt,
        "parameters": {
            "max_new_tokens": 150
        }
    }

    response = requests.post(
        MODEL_URL,
        headers=headers,
        json=payload
    )

    result = response.json()

    if "error" in result:
        return "API limit reached. Please try again later."

    return result[0]["generated_text"]


print("\nüè• Welcome to General Health Chatbot (Fast Version)")
print("Type 'exit' to stop.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Stay healthy! Goodbye üëã")
        break

    if safety_filter(user_input):
        print("\n‚ö†Ô∏è Chatbot:")
        print("This may be a serious medical situation.")
        print("Please contact emergency services immediately.\n")
        continue

    prompt = create_prompt(user_input)

    answer = generate_response(prompt)

    print("\nChatbot:", answer)
    print()
```
